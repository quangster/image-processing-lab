```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = [8, 6]
```

```python
image = cv2.imread("taj.jpg")
plt.imshow(image[...,::-1])
plt.axis("off");
```

```python
# apply bilateral filter
bilateral = cv2.bilateralFilter(
    src=image,
    d=15,
    sigmaColor=75,
    sigmaSpace=75,
)

# view the image
plt.subplot(1, 2, 1); plt.imshow(image[:,:,::-1]); plt.axis("off")
plt.subplot(1, 2, 2); plt.imshow(bilateral[:,:,::-1]); plt.axis("off")
```

```python
# apply average filter
average = cv2.blur(
    src=image,
    ksize=(5, 5),
)
plt.subplot(1, 2, 1); plt.imshow(image[:,:,::-1]); plt.axis("off")
plt.subplot(1, 2, 2); plt.imshow(average[:,:,::-1]); plt.axis("off")
```

```python
# apply median filter
median = cv2.medianBlur(
    src=image,
    ksize=5,
)
plt.subplot(1, 2, 1); plt.imshow(image[:,:,::-1]); plt.axis("off")
plt.subplot(1, 2, 2); plt.imshow(median[:,:,::-1]); plt.axis("off")
```

```python
# apply median filter
median = cv2.GaussianBlur(
    src=image,
    ksize=(5, 5),
    sigmaX=0,
)
plt.subplot(1, 2, 1); plt.imshow(image[:,:,::-1]); plt.axis("off")
plt.subplot(1, 2, 2); plt.imshow(median[:,:,::-1]); plt.axis("off")
```
